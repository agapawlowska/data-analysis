```python
import os
import glob
import math
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python

name = 'Wykres'

for filename in glob.glob('*.csv'):
    #print(str(filename))
    i=0
        
    with open(filename, 'r') as f:
        aTable = []
        bTable = []
        cTable = []
        dTable = []
        
        for line in f:
            i=i+1
            
            if i >= 23:
                a,b,c,d,e,f = line.split(',')
                a=float(a)
                b=float(b)
                c=float(c)
                d=float(d)
                aTable.append(a)
                bTable.append(b)
                cTable.append(c)
                dTable.append(d)
            
        name, rozsz = filename.split(".")
        fig = plt.figure()
        plt.title(name)
        plt.xlabel('U_in [V]')
        plt.ylabel('U_out [V]')

        plt.scatter(cTable,dTable)

        plt.show()
        
        #name, rozsz = filename.split(".")
        fig.savefig(name)
```
