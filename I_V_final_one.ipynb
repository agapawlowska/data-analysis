```python
import os
import glob
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
```

```python
#jedna kolumna danych
name = 'Wykres'

for filename in glob.glob('*.dat'):
    a = pd.read_csv(filename, sep=r"\s+")
    a.describe()
    
    name, rozsz = filename.split(".")
    #plt.title(name)
    #plt.scatter(a['voltage'], a["series1","series2"]) 
   
    fig = plt.figure()
    plt.scatter(a['voltage'], a["series1"]) 
    plt.xlabel('U_SD [V]')
    plt.ylabel('I_D [mA]')
    plt.ticklabel_format(axis='y', style='sci', scilimits=(-2,2))
    name, rozsz = filename.split(".")
    fig.savefig(name)
    

        
    
        

```

```python
#wiele kolumn danych

name = 'Wykres'

for filename in glob.glob('*.dat'):
    a = pd.read_csv(filename, sep=r"\s+")
    a.describe()
    
    name, rozsz = filename.split(".")
    #plt.title(name)
    #plt.scatter(a['voltage'], a["series1","series2"]) 
   
    fig = plt.figure()
    fig = a.plot(x='voltage', y=['series1','series2','series3','series4','series5'])
    plt.xlabel('U_SD [V]')
    plt.ylabel('I_D [mA]')
    plt.ticklabel_format(axis='y', style='sci', scilimits=(-2,2))
    name, rozsz = filename.split(".")
    fig.figure.savefig(name)
    

        
    
        

```

```python
#print(a)
b = a[['series1','series5']].mean(axis=1)
print(b)
c= a[['series1','series5']].std(axis = 1)
print(c)
```

```python
df2 = pd.DataFrame({'voltage' : a['voltage'],
                    'mean' : a[['series1','series5']].mean(axis=1), 
                    'standard_deviation' : a[['series1','series5']].std(axis = 1)})
print(df2)
```

```python
fig2 = plt.figure()
fig2 = df2.plot(x='voltage', y='mean', yerr='standard_deviation', linestyle='None',marker='.')
plt.ticklabel_format(axis='y', style='sci', scilimits=(-2,2))
plt.xlabel('U_SD [V]')
plt.ylabel('I_D [mA]')


plt.show()
```
